```python
import numpy as np
import pandas as pd

# 1. Ingestión forzada mapeando textos vacíos como nulos reales
df = pd.read_csv('hantavirus_QA_VERIFICADO.csv', na_values=["NA", "na", "NaN", "nan", " ", ""])

# 2. Renombrado corporativo de columnas
df.rename(columns={
    "cruise.crew..y.n.": "is_crew", 
    "passenger..y.n.": "is_passenger", 
    "Gh_ID": "gh_id",
    "Travel_flight": "travel_flight"
}, inplace=True)

# 3. Tratamiento de Registros Vacíos y Errores en Categóricas
df["status"] = df["status"].fillna("unspecified").astype(str).str.lower().str.strip()
df["sex"] = df["sex"].fillna("unknown").astype(str).str.lower().str.strip()
df["nationality"] = df["nationality"].fillna("unknown").astype(str).str.lower().str.strip()

# 4. Corrección de la Regla de Negocio: Hospitalización Correcta
# Solo 'confirmed' y 'probable' se hospitalizan. Los demás conservan su estado original o van a monitoreo.
is_critical = df["status"].isin(["confirmed", "probable"])
df["treatment"] = np.where(is_critical, "hospitalised", df["treatment"].astype(str).str.lower().str.strip())
df["treatment"] = np.where((df["treatment"] == "hospitalised") & (~is_critical), "monitored", df["treatment"])

# 5. Corrección de la columna Confirmed (Binario Seguro)
df["confirmed"] = np.where(is_critical, "yes", "no")

# 6. Solución Científica para la Edad (Mediana e Imputación de rango)
df["age"] = df["age"].astype(str).str.lower().str.strip()
# Si contiene 's' (como 70s), extraemos el número y le sumamos 5 (punto medio de la década)
has_s = df["age"].str.contains("s", na=False)
df["age"] = pd.to_numeric(df["age"].str.extract(r"(\d+)", expand=False), errors="coerce")
df.loc[has_s, "age"] += 5

# Calcular mediana de la edad para registros vacíos
mediana_edad = df["age"].median()
df["age"].fillna(round(mediana_edad), inplace=True)
df["age"] = df["age"].astype("Int64")

# 7. Imputación Temporal (Incubación de 18 días)
df["ship_boarded"] = pd.to_datetime("2026-04-01")
df["symptom_onset"] = pd.to_datetime(df["symptom_onset"], errors="coerce")
df["symptom_onset"].fillna(df["ship_boarded"] + pd.Timedelta(days=18), inplace=True)

# 8. Desglose atómico de síntomas (One-Hot Encoding)
df["symptoms"] = df["symptoms"].fillna("no symptoms").astype(str).str.lower().str.strip()
df["symptoms"].replace({"diarreah": "diarrhea", "diarrhiea": "diarrhea"}, regex=True, inplace=True)

all_symptoms = df["symptoms"].str.split(",").explode().str.strip()
unique_symptoms = all_symptoms[(all_symptoms != "") & (all_symptoms != "no symptoms") & (~all_symptoms.str.contains(r"\(", na=False))].unique()

for symptom in unique_symptoms:
    col_name = f"has_{symptom.replace(' ', '_').replace('-', '_')}"
    df[col_name] = df["symptoms"].str.contains(symptom, regex=False).astype(int)

# 9. Eliminación final de columnas con ruido de alta cardinalidad
columnas_ruido = ["linked_information", "context", "sources", "accession_id", "symptoms"]
df.drop(columns=columnas_ruido, errors="ignore", inplace=True)

# 10. Formateo de IDs remanentes
df["linked_id"] = pd.to_numeric(df["linked_id"], errors="coerce").fillna(0).astype(int)

# Guardado definitivo
df.to_csv("hantavirus_PRODUCTION_READY.csv", index=False)

print("🚀 ¡DATASET BLINDADO Y PERSISTIDO EN DISCO!")
print(f"Dimensiones finales: {df.shape[0]} filas x {df.shape[1]} columnas.")
print(f"Columnas limpias listas para BI: {list(df.columns)}")
```

```python
def apply_cause_death_protocol(file_path: str) -> pd.DataFrame:
    # 1. Ingestión del dataset mapeando nulos iniciales
    df = pd.read_csv(file_path, na_values=["NA", "na", "NaN", "nan", " ", ""], keep_default_na=True)
    
    # 2. Estandarización de columnas clave para el cruce lógico
    df["outcome"] = df["outcome"].astype(str).str.lower().str.strip().replace("nan", np.nan)
    df["cause_death"] = df["cause_death"].astype(str).str.lower().str.strip().replace("nan", np.nan)
    
    # =============================================================================
    # REGLA DE NEGOCIO CLÍNICA: IMPUTACIÓN DE CAUSA DE MUERTE POR HANTAVIRUS
    # =============================================================================
    # Condición A: El paciente falleció (outcome == 'died')
    has_died = df["outcome"] == "died"
    
    # Condición B: La causa de muerte está en blanco (isnull)
    cause_is_blank = df["cause_death"].isnull()
    
    # Aplicamos el protocolo epidemiológico:
    # Si el paciente falleció y la causa estaba en blanco, asignamos el Síndrome Pulmonar.
    # Si ya tenía un registro previo (como 'respiratory distress'), lo preservamos.
    df["cause_death"] = np.where(
        has_died & cause_is_blank, 
        "hantavirus pulmonary syndrome", 
        df["cause_death"]
    )
    
    # Para los pacientes que NO fallecieron, nos aseguramos de limpiar la columna 
    # asignando una etiqueta de control limpia ("not applicable") para evitar NaN molestos en Power BI.
    df["cause_death"] = np.where(
        ~has_died, 
        "not applicable", 
        df["cause_death"]
    )

    # =============================================================================
    # Bloque de Consolidación y Limpieza Radical (Sprints Anteriores)
    # =============================================================================
    df.rename(columns={"cruise.crew..y.n.": "is_crew", "passenger..y.n.": "is_passenger", "Gh_ID": "gh_id"}, inplace=True)
    df["status"] = df["status"].fillna("unspecified").astype(str).str.lower().str.strip()
    
    # Corrección de Hospitalización y Flags
    is_critical = df["status"].isin(["confirmed", "probable"])
    df["treatment"] = np.where(is_critical, "hospitalised", df["treatment"].astype(str).str.lower().str.strip())
    df["confirmed"] = np.where(is_critical, "yes", "no")
    
    # Edad a Entero Puro (Int64)
    df["age"] = df["age"].astype(str).str.lower().str.strip()
    has_s = df["age"].str.contains("s", na=False)
    df["age"] = pd.to_numeric(df["age"].str.extract(r"(\d+)", expand=False), errors="coerce")
    df.loc[has_s, "age"] += 5
    df["age"].fillna(round(df["age"].median()), inplace=True)
    df["age"] = df["age"].astype("Int64")
    
    # Desglose de síntomas (One-Hot Encoding)
    df["symptoms"] = df["symptoms"].fillna("no symptoms").astype(str).str.lower().str.strip()
    all_symptoms = df["symptoms"].str.split(",").explode().str.strip()
    unique_symptoms = all_symptoms[(all_symptoms != "") & (all_symptoms != "no symptoms") & (~all_symptoms.str.contains(r"\(", na=False))].unique()
    for symptom in unique_symptoms:
        df[f"has_{symptom.replace(' ', '_').replace('-', '_')}"] = df["symptoms"].str.contains(symptom, regex=False).astype(int)
        
    # Eliminación de ruido y guardado persistente
    columnas_ruido = ["linked_information", "context", "sources", "accession_id", "symptoms"]
    df.drop(columns=columnas_ruido, errors="ignore", inplace=True)
    
    return df

# Ejecución y Auditoría Médica en tiempo real
if __name__ == "__main__":
    PATH_INPUT = "hantavirus_QA_VERIFICADO.csv"
    df_finalizado = apply_cause_death_protocol(PATH_INPUT)
    
    print("📋 ### AUDITORÍA DE PACIENTES FALLECIDOS (POST-REGLA) ###")
    # Filtramos para ver cómo quedaron los registros de fallecidos
    df_fallecidos = df_finalizado[df_finalizado["outcome"] == "died"]
    print(df_fallecidos[["gh_id", "status", "outcome", "cause_death"]].to_string(index=False))
    
    print("\n" + "="*70 + "\n")
    print("📊 ### CONTEO DE CATEGORÍAS EN CAUSE_DEATH ###")
    print(df_finalizado["cause_death"].value_counts())
    
    # Guardar los cambios físicos de forma definitiva
    df_finalizado.to_csv("hantavirus_PRODUCTION_READY.csv", index=False)
    print("\n Archivo 'hantavirus_PRODUCTION_READY.csv' actualizado y guardado.")
```

```python
def transform_symptoms_to_text(file_path: str) -> pd.DataFrame:
    # 1. Cargamos el dataset que ya incluye el procesamiento previo
    # (Asegúrate de apuntar al archivo que generamos en el paso anterior)
    df = pd.read_csv(file_path)
    
    # =============================================================================
    # TRANSFORMAR COLUMNAS 'has_' DE BINARIO (1/0) A TEXTO (yes/no)
    # =============================================================================
    # A. Identificamos de forma dinámica todas las columnas que empiezan con 'has_'
    symptom_columns = [col for col in df.columns if col.startswith("has_")]
    
    print(f"📢 Columnas de síntomas detectadas para transformación: {symptom_columns}")
    
    # B. Iteramos sobre cada columna detectada para hacer la sustitución
    for col in symptom_columns:
        # Reemplazamos 1 por 'yes' y 0 por 'no'
        # Usamos .map() o np.where() asegurando que se guarden como tipo texto (str)
        df[col] = df[col].map({1: "yes", 0: "no"}).astype(str)
    
    return df

# Ejecución y control de calidad inmediato
if __name__ == "__main__":
    # Usamos el último archivo generado en la sesión
    PATH_INPUT = "hantavirus_PRODUCTION_READY.csv" 
    
    # Ejecutamos la transformación
    df_visualizacion = transform_symptoms_to_text(PATH_INPUT)
    
    print("\n" + "="*70 + "\n")
    print("📋 ### AUDITORÍA DE FORMATO CATEGÓRICO (Primeros 5 registros) ###")
    
    # Busquemos un par de columnas de síntomas para mostrar el control de calidad
    symptom_cols = [col for col in df_visualizacion.columns if col.startswith("has_")]
    columns_to_show = ["gh_id", "status"] + symptom_cols[:4] # Mostramos los primeros 4 síntomas
    
    print(df_visualizacion[columns_to_show].head(5).to_string(index=False))
    
    # =============================================================================
    # GUARDADO DEL ARCHIVO FINAL DE PRODUCCIÓN
    # =============================================================================
    # Mantenemos el mismo nombre para consolidar toda la jerarquía de limpieza
    df_visualizacion.to_csv("hantavirus_PRODUCTION_READY.csv", index=False)
    print("\n✅ ¡Sustitución completada con éxito!")
    print("💾 El archivo 'hantavirus_PRODUCTION_READY.csv' ha sido actualizado y está 100% optimizado para tus dashboards.")
```

```python
def apply_epidemiological_incubation_protocol(file_path: str) -> pd.DataFrame:
    # 1. Ingestión limpia del dataset
    df = pd.read_csv(file_path, na_values=["NA", "na", "NaN", "nan", " ", ""])
    
    # 2. Estandarización base de nombres y estados preliminares
    df.rename(columns={
        "cruise.crew..y.n.": "is_crew", 
        "passenger..y.n.": "is_passenger", 
        "Gh_ID": "gh_id",
        "Travel_flight": "travel_flight"
    }, inplace=True)
    
    df["status"] = df["status"].fillna("unspecified").astype(str).str.lower().str.strip()
    
    # 3. Forzar el formato Datetime en variables temporales clave
    df["ship_boarded"] = pd.to_datetime("2026-04-01") # Fecha base constante del brote
    df["left_ship"] = pd.to_datetime(df["left_ship"], errors="coerce")
    df["symptom_onset"] = pd.to_datetime(df["symptom_onset"], errors="coerce")

    # =============================================================================
    # IMPUTACIÓN EPIDEMIOLÓGICA DE REGISTROS VACÍOS EN LEFT_SHIP
    # =============================================================================
    # Si la fecha en que bajó del barco está vacía, asumimos epidemiológicamente 
    # que completó el periodo de aislamiento preventivo de la cohorte (finales de mayo/Tenerife)
    # o la fecha de inicio de síntomas si ocurrió a bordo.
    df["left_ship"].fillna(pd.to_datetime("2026-05-10"), inplace=True)

    # 4. Cálculo matemático de los días de exposición a bordo
    df["days_exposed"] = (df["left_ship"] - df["ship_boarded"]).dt.days
    
    # =============================================================================
    # REGLA DE FILTRADO DE SOSPECHA Y OBSERVACIÓN POR INCUBACIÓN
    # =============================================================================
    # Analizamos si el paciente presenta síntomas a través del texto original antes de dropearlo
    df["symptoms"] = df["symptoms"].fillna("no symptoms").astype(str).str.lower().str.strip()
    has_active_symptoms = (df["symptoms"] != "no symptoms") & (df["symptoms"] != "")
    
    # Definimos quiénes están en un limbo de diagnóstico (no son confirmados, ni probables, ni negativos)
    in_limbo = df["status"].isin(["monitored", "tested", "inconclusive", "unspecified"])
    
    # Aplicación de las subcategorías epidemiológicas en la columna status:
    # Caso A: En el limbo + Estuvo expuesto + SÍNTOMAS ACTIVOS => Sospecha Activa
    df["status"] = np.where(
        in_limbo & (df["days_exposed"] > 0) & has_active_symptoms,
        "active suspicion",
        df["status"]
    )
    
    # Caso B: En el limbo + Estuvo expuesto + SIN SÍNTOMAS => En Observación Clínica
    df["status"] = np.where(
        in_limbo & (df["days_exposed"] > 0) & (~has_active_symptoms),
        "under clinical observation",
        df["status"]
    )

    # =============================================================================
    # Consolidación de Procesos de Sprints Anteriores (Gobernanza Completa)
    # =============================================================================
    # Corrección de Hospitalización y Confirmación basadas en el nuevo mapa epidemiológico
    is_critical = df["status"].isin(["confirmed", "probable", "active suspicion"])
    df["treatment"] = np.where(is_critical, "hospitalised", df["treatment"].astype(str).str.lower().str.strip())
    df["treatment"] = np.where(df["status"] == "under clinical observation", "quarantine isolation", df["treatment"])
    df["confirmed"] = np.where(df["status"] == "confirmed", "yes", "no")
    
    # Causa de muerte condicional
    df["outcome"] = df["outcome"].astype(str).str.lower().str.strip().replace("nan", np.nan)
    df["cause_death"] = np.where(df["outcome"] == "died", "hantavirus pulmonary syndrome", "not applicable")
    
    # Formateo de Edad (Int64)
    df["age"] = df["age"].astype(str).str.lower().str.strip()
    has_s = df["age"].str.contains("s", na=False)
    df["age"] = pd.to_numeric(df["age"].str.extract(r"(\d+)", expand=False), errors="coerce")
    df.loc[has_s, "age"] += 5
    df["age"].fillna(round(df["age"].median()), inplace=True)
    df["age"] = df["age"].astype("Int64")
    
    # Variables geográficas y vuelos
    df["travel_from"] = df["travel_from"].fillna("not reported").astype(str).str.strip()
    df["travel_to"] = df["travel_to"].fillna("not reported").astype(str).str.strip()
    df["nationality"] = df["nationality"].fillna("unknown").astype(str).str.lower().str.strip()
    df["left_location"] = df["left_location"].fillna("not specified").astype(str).str.strip()
    df["travel_flight"] = df["travel_flight"].fillna("charter/unspecified").astype(str).str.strip()
    df["linked_id"] = pd.to_numeric(df["linked_id"], errors="coerce").fillna(0).astype("Int64")
    
    # Desglose de síntomas dinámico a categórico ("yes"/"no")
    df["symptoms"].replace({"diarreah": "diarrhea", "diarrhiea": "diarrhea"}, regex=True, inplace=True)
    all_symptoms = df["symptoms"].str.split(",").explode().str.strip()
    unique_symptoms = all_symptoms[(all_symptoms != "") & (all_symptoms != "no symptoms") & (~all_symptoms.str.contains(r"\(", na=False))].unique()
    for symptom in unique_symptoms:
        df[f"has_{symptom.replace(' ', '_').replace('-', '_')}"] = np.where(df["symptoms"].str.contains(symptom, regex=False), "yes", "no")
        
    # Eliminación de columnas de texto libre originales (Ruido)
    columnas_basura = ["linked_information", "context", "sources", "accession_id", "symptoms"]
    df.drop(columns=columnas_basura, errors="ignore", inplace=True)
    
    return df

if __name__ == "__main__":
    df_epidemiologico = apply_epidemiological_incubation_protocol("hantavirus_QA_VERIFICADO.csv")
    
    print("📊 ### NUEVA SEGMENTACIÓN DE CONTROL SANITARIO (STATUS) ###")
    print(df_epidemiologico["status"].value_counts())
    
    print("\n" + "="*70 + "\n")
    print("📋 ### REVISIÓN DE PACIENTES BAJO NUEVAS REGLAS CLÍNICAS ###")
    print(df_epidemiologico[["gh_id", "status", "days_exposed", "treatment"]].head(10).to_string(index=False))
    
    # Mantenemos el mismo nombre para consolidar toda la jerarquía de limpieza
    df_visualizacion.to_csv("hantavirus_PRODUCTION_READY.csv", index=False)
    print("\n✅ ¡Sustitución completada con éxito!")
    print("💾 El archivo 'hantavirus_PRODUCTION_READY.csv' ha sido actualizado y está 100% optimizado para tus dashboards.")
```

```python
# 1. Leer el dataset de producción actual
df = pd.read_csv('hantavirus_PRODUCTION_READY.csv')

# 2. Agrupar las variantes ortográficas detectadas en la auditoría
columnas_variantes = ['has_diarrhea', 'has_diarreah', 'has_diarrhiea']

# 3. Consolidación vectorial (Si cualquiera tiene "yes", la correcta será "yes")
df['has_diarrhea'] = np.where(
    (df[columnas_variantes] == 'yes').any(axis=1), 
    'yes', 
    'no'
)

# 4. Eliminar definitivamente las columnas con errores ortográficos
df.drop(columns=['has_diarreah', 'has_diarrhiea'], errors='ignore', inplace=True)

# 5. Guardado forzoso para asegurar la persistencia en el archivo físico
df.to_csv('hantavirus_PRODUCTION_READY.csv', index=False)

print("✅ ¡Estandarización de 'has_diarrhea' completada y guardada en disco!")
```

```python

# 1. Carga limpia del archivo de producción centralizado
archivo_central = 'hantavirus_PRODUCTION_READY.csv'
df = pd.read_csv(archivo_central, na_values=["NA", "na", "NaN", "nan", " ", ""])

# 2. Corrección de nombres de columnas logísticas
if "Travel_flight" in df.columns:
    df.rename(columns={"Travel_flight": "travel_flight"}, inplace=True)

# 3. Homogeneización de Flags Binarios al estándar corporativo "yes"/"no"
df["is_crew"] = df["is_crew"].map({"y": "yes", "n": "no", "yes": "yes", "no": "no"}).fillna("no")
df["is_passenger"] = df["is_passenger"].map({"y": "yes", "n": "no", "yes": "yes", "no": "no"}).fillna("no")

# 4. Tratamiento Avanzado de Identificadores Cruzados (Formato Entero Puro)
df["linked_id"] = pd.to_numeric(df["linked_id"], errors="coerce").fillna(0).astype("Int64")

# 5. Estandarización y Normalización de Variables Geográficas y de Vuelo
df["travel_from"] = df["travel_from"].fillna("not reported").astype(str).str.strip()
df["travel_to"] = df["travel_to"].fillna("not reported").astype(str).str.strip()
df["left_location"] = df["left_location"].fillna("not specified").astype(str).str.strip()
df["nationality"] = df["nationality"].fillna("unknown").astype(str).str.lower().str.strip()

# Limpieza de strings en aerolíneas e imputación asistida por rol corporativo
df["travel_flight"] = df["travel_flight"].astype(str).str.strip().replace("nan", np.nan)
df["travel_flight"] = np.where((df["is_crew"] == "yes") & (df["travel_flight"].isnull()), "crew member logistics", df["travel_flight"])
df["travel_flight"].fillna("charter/unspecified", inplace=True)

# 6. Casteo Forzoso de Variables Temporales (Fechas Nativas para BI)
df["ship_boarded"] = pd.to_datetime("2026-04-01")
df["left_ship"] = pd.to_datetime(df["left_ship"], errors="coerce")
# Regla Epidemiológica: Imputación por fin de aislamiento de la cohorte
df["left_ship"].fillna(pd.to_datetime("2026-05-10"), inplace=True)

for col in ["symptom_onset", "outcome_date", "treatment_date", "confirmation_date"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Cálculo de la métrica de exposición activa
df["days_exposed"] = (df["left_ship"] - df["ship_boarded"]).dt.days

# 7. Remoción de columnas basura de texto libre heredadas para mitigar ruido en BI
columnas_ruido = ["linked_information", "context", "sources", "accession_id"]
df.drop(columns=columnas_ruido, errors="ignore", inplace=True)

# =============================================================================
# GUARDADO PERSISTENTE SOBRE EL MISMO ARCHIVO ORIGEN
# =============================================================================
df.to_csv(archivo_central, index=False)

print(f"✅ ¡Dataset '{archivo_central}' completamente optimizado y actualizado de forma directa!")
print(f"Estructura final consolidada: {df.shape[0]} filas x {df.shape[1]} columnas.")
```

```python
# 1. Cargar el dataset de producción central
archivo_central = 'hantavirus_PRODUCTION_READY.csv'
df = pd.read_csv(archivo_central)

# 2. Forzar conversión a tipo datetime para operar matemáticamente
df['confirmation_date'] = pd.to_datetime(df['confirmation_date'], errors='coerce')

# 3. Calcular la media matemática de la línea de tiempo (ignoring NaNs)
timestamps = df['confirmation_date'].dropna().astype('int64')
media_timestamp = timestamps.mean()
fecha_media = pd.to_datetime(media_timestamp)

# 4. Imputar vacíos con la fecha promedio exacta calculada (2026-05-08)
df['confirmation_date'].fillna(pd.to_datetime(fecha_media.date()), inplace=True)

# 5. Volver a formatear a string limpio YYYY-MM-DD para su almacenamiento estandarizado
df['confirmation_date'] = df['confirmation_date'].dt.strftime('%Y-%m-%d')

# 6. Guardar los cambios directamente en el mismo archivo físico
df.to_csv(archivo_central, index=False)

print("✅ ¡Imputación temporal por fecha media completada en 'confirmation_date'!")
```

```python
# 1. Cargar el dataset único de producción
archivo_central = 'hantavirus_PRODUCTION_READY.csv'
df = pd.read_csv(archivo_central)

# 2. Convertir las columnas de tiempo a formato datetime nativo
df['ship_boarded'] = pd.to_datetime(df['ship_boarded'])
df['treatment_date'] = pd.to_datetime(df['treatment_date'], errors='coerce')

# 3. IMPUTACIÓN CONDICIONAL BAJO REGLA DE INCUBACIÓN CRÍTICA (7 DÍAS)
# Caso A: Si el paciente sobrevivió, recibió tratamiento oportuno (Día 4 del brote: 2026-04-05)
df.loc[df['treatment_date'].isnull() & (df['outcome'] != 'died'), 'treatment_date'] = pd.to_datetime('2026-04-05')

# Caso B: Si el paciente falleció, el tratamiento llegó fuera de la ventana de 7 días (2026-04-12)
df.loc[df['treatment_date'].isnull() & (df['outcome'] == 'died'), 'treatment_date'] = pd.to_datetime('2026-04-12')

# 4. Re-formatear la columna al estándar limpio de producción YYYY-MM-DD
df['treatment_date'] = df['treatment_date'].dt.strftime('%Y-%m-%d')

# 5. Sobreescribir los cambios directamente sobre el archivo físico original
df.to_csv(archivo_central, index=False)

print("✅ ¡Sincronización analítica entre tratamiento y causa de muerte completada con éxito!")
```

```python
# 1. Cargar el dataset único de producción
archivo_central = 'hantavirus_PRODUCTION_READY.csv'
df = pd.read_csv(archivo_central)

# 2. Convertir fechas y calcular la columna temporal de días transcurridos hasta el tratamiento
df['ship_boarded'] = pd.to_datetime(df['ship_boarded'])
df['treatment_date'] = pd.to_datetime(df['treatment_date'])
days_to_treatment = (df['treatment_date'] - df['ship_boarded']).dt.days

# 3. Sumar de forma vectorial la presencia de síntomas de alta severidad
sintomas_criticos = ['has_fever', 'has_shortness_of_breath', 'has_pneumonia', 'has_respiratory_symptoms']
df_sintomas_bool = df[sintomas_criticos].replace({'yes': True, 'no': False})
num_critical_symptoms = df_sintomas_bool.sum(axis=1)

# 4. CONDICIÓN MATEMÁTICA: Si es un caso activo, con >= 2 síntomas críticos y tratamiento tardío (> 7 días)
condicion_fallece = (df['outcome'].isnull()) & (days_to_treatment > 7) & (num_critical_symptoms >= 2) & (df['status'].isin(['confirmed', 'probable']))

# Aplicar asignaciones en los registros vacíos
df.loc[condicion_fallece, 'outcome'] = 'died'
df['outcome'].fillna('alive', inplace=True)

# 5. Sobreescribir los cambios de forma forzosa sobre el archivo físico original
df.to_csv(archivo_central, index=False)

print("✅ ¡Clasificación predictiva de desenlace clínico completada!")
```
